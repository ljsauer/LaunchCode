```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Seaborn Themes

# Global configuration settings
sns.set(context='notebook', style='darkgrid', palette='dark')
```

```python
# Built-in color palettes
sns.palplot(sns.color_palette())
sns.palplot(sns.color_palette('husl', 8))
sns.palplot(sns.color_palette('hls', 8))
```

```python
# Matplotlib colormap of evenly-spaced colors
sns.palplot(sns.color_palette('coolwarm', 7))
```

```python
# Sequential palette with linear increase in brightness
sns.palplot(sns.cubehelix_palette(8))
```

```python
# Palettes are used in a plot via the color parameter:
gammas = sns.load_dataset('gammas')
sns.tsplot(gammas, 'timepoint', 'subject', 'ROI', 'BOLD signal',
          color='muted')
```

```python
# Advanced Plots in Seaborn

# facetting histograms by subsets of data
sns.set(style='darkgrid')

tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips, row='sex', col='time', margin_titles=True)
bins = np.linspace(0, 60, 13)
g.map(plt.hist, 'total_bill', color='steelblue', bins=bins, lw=0)
```

```python
# Distribution plot examples

sns.set(style='white', palette='muted')
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)

rs = np.random.RandomState(10)

b, g, r, p = sns.color_palette('muted', 4)

d = rs.normal(size=100)

sns.distplot(d, kde=False, color=b, ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color=r, ax=axes[0, 1])
sns.distplot(d, hist=False, color=g, kde_kws={'shade':True}, ax=axes[1, 0])
sns.distplot(d, color=p, ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```

```python
# hexbin plot with marginal distributions

from scipy.stats import kendalltau
sns.set(style='ticks')

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = (-0.5 * x) + rs.normal(size=1000)
sns.jointplot(x, y, kind='hex', stat_func=kendalltau, color='#4CB391')
```

```python
# faceted logistic regression

sns.set(style='darkgrid')
df = sns.load_dataset('titanic')

pal = dict(male='#6495ED', female='#F08080')
g = sns.lmplot('age', 'survived', col='sex', hue='sex', data=df,
              palette=pal, y_jitter=0.02, logistic=True)

g.set(xlim=(0, 80), ylim=(-0.05, 1.05))
```

```python
# Linear regression with marginal distribution

sns.set(style='darkgrid')
tips = sns.load_dataset('tips')
color = sns.color_palette()[2]
g = sns.jointplot('total_bill', 'tip', data=tips, kind='reg',
                 xlim=(0, 60), ylim=(0, 12), color=color, size=7)
```

```python
# Time series Catplot

sns.set(style='white')
planets = sns.load_dataset('planets')
years = np.arange(2000, 2015)
g = sns.catplot('year', data=planets, palette='BuPu',
                  kind='count', aspect=1.5)
g.set_xticklabels(step=2)
```

```python
# Pair plot

sns.set(style='darkgrid')
df = sns.load_dataset('iris')
sns.pairplot(df, hue='species', height=2.5)
```
