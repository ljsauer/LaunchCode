```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Seaborn Themes

# Global configuration settings
sns.set(context='notebook', style='darkgrid', palette='dark')
```

```python
# Built-in color palettes
sns.palplot(sns.color_palette())
sns.palplot(sns.color_palette('husl', 8))
sns.palplot(sns.color_palette('hls', 8))
```

```python
# Matplotlib colormap of evenly-spaced colors
sns.palplot(sns.color_palette('coolwarm', 7))
```

```python
# Sequential palette with linear increase in brightness
sns.palplot(sns.cubehelix_palette(8))
```

```python
# Palettes are used in a plot via the color parameter:
gammas = sns.load_dataset('gammas')
sns.tsplot(gammas, 'timepoint', 'subject', 'ROI', 'BOLD signal',
          color='muted')
```

```python
# Advanced Plots in Seaborn

# facetting histograms by subsets of data
sns.set(style='darkgrid')

tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips, row='sex', col='time', margin_titles=True)
bins = np.linspace(0, 60, 13)
g.map(plt.hist, 'total_bill', color='steelblue', bins=bins, lw=0)
```

```python
# Distribution plot examples

sns.set(style='white', palette='muted')
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)

rs = np.random.RandomState(10)

b, g, r, p = sns.color_palette('muted', 4)

d = rs.normal(size=100)

sns.distplot(d, kde=False, color=b, ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color=r, ax=axes[0, 1])
sns.distplot(d, hist=False, color=g, kde_kws={'shade':True}, ax=axes[1, 0])
sns.distplot(d, color=p, ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
