```python
import pandas as pd

digits = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra",
                     header=None)
digits.describe()
```

```python
first = digits.head(5)
last = digits.tail(5)

print(first)
print(last)
```

```python
# for large datasets, it might help to take a sample of your data
# as an easy way to get a feel for it quickly
digits.sample(5)
```

```python
# another more complicated way to do this is by creating a random index
# and then getting random rows from your DataFrame
from random import sample
import numpy as np

randomIndex = np.array(sample(range(len(digits)), 5))

digitsSample = digits.iloc[randomIndex]

print(digitsSample)
```

```python
# Identifying and filling missing values
pd.isnull(digits)
```
