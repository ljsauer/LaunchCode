```python
"""
Tutorial #2

https://www.datacamp.com/community/tutorials/python-data-profiling
"""
import pandas as pd

beers = pd.read_csv('data/processed/beers.csv')
breweries = pd.read_csv('data/processed/breweries.csv')

bnbs = pd.merge(beers, breweries, how='inner', left_on='brewery_id', right_on='id', sort=True,
               suffixes=('_beer', '_brewery'))
```

```python
# build a function to group float64 and int64 data types
def get_var_category(series):
    unique_count = series.nunique(dropna=False)
    total_count = len(series)
    if series.dtype == float or series.dtype == int:
        return 'Numerical'
    elif unique_count == total_count:
        return 'Text (Unique)'
    else:
        return 'Categorical'

def print_categories(df):
    for column_name in df.columns:
        print(column_name, ': ', get_var_category(df[column_name]))
```

```python
print_categories(beers)
```

```python
print_categories(breweries)
```

```python
# find missing values of a variable
length_ibu = len(beers['ibu'])
count_ibu = beers['ibu'].count()

missing_values_ibu = length_ibu - count_ibu
pct_missing_values_ibu = float(missing_values_ibu / length_ibu)
pct_missing_values_ibu = '{0:.1f}%'.format(pct_missing_values_ibu * 100)
print(pct_missing_values_ibu)
```

```python
# find the minimum and maximum values of the ibu variable to understand the range
min_val_ibu = beers['ibu'].min()
max_val_ibu = beers['ibu'].max()

print('Minimun value: ', min_val_ibu,', ', 'Maximum value: ', max_val_ibu)
```

```python
# find the mode of the ibu variable
ibu_mode = beers['ibu'].mode()

print(ibu_mode)
```

```python
# find the mean of the ibu variable -- the measure of central tendency
ibu_mean = beers['ibu'].mean()

print(ibu_mean)
```

```python
# find the ibu median -- sometimes a better measure of central tendency than the mean
ibu_median = beers['ibu'].median()

print(ibu_median)
```

```python
# find the standard deviation of the ibu variable to measure dispersion
ibu_stdev = beers['ibu'].std()

print(ibu_stdev)
```

```python
# find the quantile stats of the ibu variable
quantile_ibu = beers['ibu'].quantile([.25, .5, .75])

print(quantile_ibu)
```

```python
# make a distribution plot to visualize the distribution of values in the ibu variable
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
sns.set_palette(sns.color_palette('muted'))

sns.distplot(beers['ibu'].dropna())
```
