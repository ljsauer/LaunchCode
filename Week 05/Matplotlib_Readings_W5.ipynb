```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Cleaning up plots -- this:
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

```python
# . . .as opposed to this:
plt.plot([1, 2, 3, 4], [10, 20, 24, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

```python
# Creating a plot

# initialize a figure
fig = plt.figure()

# add axes to the figure
fig.add_axes([0,0,1,1])
```

```python
# Subplots -- set-up and place your axes on a regular grid
fig = plt.figure()

# set up axes
ax = fig.add_subplot(111)
# 111 is equal to 1,1,1 -- this is actually three arguments designating the
# number of rows, columns, and the plot number (1 subplot)

# scatter the data
ax.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

plt.show()
```

```python
""" 
Note: add_axes() & add_subplot() only differ in that the former allows for
control over the positioning of the axes, while the latter simply allows
the axes to be situated according to a subplot grid.
"""
```

```python
# Changing the size of figures
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
"""OR: <fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))>"""

# Plot the data
ax1.bar([1, 2, 3],[3, 4, 5])
ax2.barh([0.5, 1, 2.5],[0, 1, 2])

plt.show
```
